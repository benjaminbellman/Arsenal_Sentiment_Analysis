```python
import snscrape.modules.twitter as sntwitter
import pandas as pd

### Importing DataFrame of Gamedays and formatting date Properly
df_queries = pd.read_csv(r'..\..\..\OneDrive\Desktop\Twitter\Arsenal_Tottenham_Games_Twitter.csv')

## Formatting Date
df_queries['Date'] = pd.to_datetime(df_queries['Date'])
df_queries['Date'] = pd.to_datetime(df_queries['Date'], format = '%yyyy-%mm-%dd')
df_queries['Date'] = df_queries['Date'].apply(str)
df_queries['Date'] = df_queries['Date'].str[:-9]

### Creating combined query column and converting to list. 
df_queries['combined'] = df_queries[['Team','Date']].values.tolist()
query_list = list(df_queries['combined'])

## Defining get_all_queries
def get_queries():
    queries = []
    for value in query_list:
        query = "{} until:{}".format(value[0],value[1])
        queries.append(query)
    return queries

all_queries = get_queries()


## Defining our get_tweets function

def get_tweets(query,limit =2): 
    tweets = []
    for tweet in sntwitter.TwitterSearchScraper(query).get_items():
        #print(vars(tweet))
        #break
        if len(tweets) == limit:
            break
        else:
            tweets.append([tweet.date, 
                           tweet.user.username, 
                           tweet.content, 
                           tweet.likeCount, 
                           tweet.replyCount, 
                           tweet.retweetCount])
    return tweets

records = []
for query in all_queries[0:100]:
    records.extend(get_tweets(query))
```

```python
df_tweets = pd.DataFrame(records,columns =['Date','User','Tweet','TweetLikes','TweetReplies','RetweetCount']) 
print(df_tweets.shape)
```

```python
df_queries = pd.DataFrame(large_list,columns =['Query']) 
```
